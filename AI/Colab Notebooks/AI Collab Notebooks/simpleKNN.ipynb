```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np




```

```python
#Utility functions 

def calcDistance(df, unlabeled):
  d = []
  for i in range (len(df)):
    x1 = df.iloc[i].x
    y1 = df.iloc[i].y
    p2 = np.array((x1, y1))
    dx = np.sqrt(np.sum(np.square(unlabeled - p2)))
    d.append(dx.round(2))
  df['distance'] = d
  return df

def calcKNN(data, knn):
  data.sort_values(by=['distance'], ascending=True, inplace=True)
  data_knn = data.head(knn)
  p = data_knn['class'].value_counts()
  return p
```

```python
#runner
def main(x,y,c,t_x, t_y):
  status = True

 # create dataframe with values
  
  dataframe = pd.DataFrame(zip(x,y,c),columns= column_names)
  print(dataframe)


  #Step 1: choose value for K
  k = 3 
  
  #Step 2: calculate the Eucledian distance from target
  target = [t_x, t_y]
  answer = calcDistance(dataframe, target)
  #Step 3: Determine nearest neighbor
  nn = calcKNN(answer,k)
  print(nn)

  #Step 4: predict the class label based on the majority of votes
  return status
```

```python
#set up data sets
x = [4,5,6,7,4,5,6,7,4,5,6,7]
y = [3,4,4,3,12,14,10,10,6,7,7,6]
#labels or classes
c = ['soccer','soccer','soccer','soccer', 'Gym','Gym','Gym','Gym',"Bas","Bas","Bas","Bas"]
column_names = ["Height", "Speed", "Class"]

#Target we want to predict 

target_1_x = 4
target_1_y = 3

s = main(x,y,c, target_1_x, target_1_y)
```
